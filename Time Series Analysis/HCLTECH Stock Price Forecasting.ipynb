```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline
sns.set(rc={
    'figure.figsize':(15, 10)
})
```

```python
df=pd.read_csv(r'D:\HCLTECH.csv')
df.head(10)
```

```python
df.shape
```

```python
df.drop(df.index[0:4549], inplace=True)
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.set_index(pd.to_datetime(df['Date']), inplace=True)
df.head()
```

```python
df.shape
```

```python
df=df[['Prev Close']]
df.head()
```

```python
df=df.asfreq('W', method='pad')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.plot();
plt.show()
```

```python
plt.plot(df, label='Original')
plt.plot(df.rolling(window=5).mean(), color='red', label='Rolling Avg')
plt.plot(df.rolling(window=5).std(), color='green', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Average and Standard Deviation')
plt.tight_layout()
plt.show()
```

```python
plt.plot(np.log(df))
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
mul=seasonal_decompose(np.log(df), model='multiplicative', period=1)
mul.plot().suptitle('\nMultiplicative Decompose', fontsize=12);
plt.tight_layout()
plt.show()
```

```python
log_shift_diff=np.log(df)-np.log(df.shift(1))
log_shift_diff.dropna(inplace=True)
plt.plot(log_shift_diff, label='Original')
plt.plot(log_shift_diff.rolling(window=12).mean(), color='red', label='Rolling Avg')
plt.plot(log_shift_diff.rolling(window=12).std(), color='green', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Average and Standard Deviation')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
ACF=acf(log_shift_diff, nlags=15)
PACF=pacf(log_shift_diff, nlags=15)
```

```python
plt.plot(ACF, marker='+')
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(log_shift_diff)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(log_shift_diff)), linestyle='--', color='gray')
plt.title('Auto Correlation')
plt.xlabel('Lags')
plt.ylabel('Correlation')
plt.tight_layout()
plt.show()
```

```python
plt.plot(PACF, marker='+')
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(log_shift_diff)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(log_shift_diff)), linestyle='--', color='gray')
plt.title('Partial Auto Correlation')
plt.xlabel('Lags')
plt.ylabel('Correlation')
plt.tight_layout()
plt.show()
```

```python
fig=plt.figure(figsize=(12, 8))
ax1=fig.add_subplot(211)
fig=plot_acf(log_shift_diff.dropna(), lags=15, ax=ax1)
ax2=fig.add_subplot(212)
fig=plot_pacf(log_shift_diff.dropna(), lags=15, ax=ax2)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model1=ARIMA(df, order=(1, 1, 1))
results1=model1.fit(disp=False)
print(results1.summary())
plt.plot(df)
plt.plot(results1.fittedvalues, color='red')
plt.show()
```

```python
import statsmodels.api as sm
model2=sm.tsa.statespace.SARIMAX(df, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results2=model2.fit()
print(results2.summary())
```

```python
plt.plot(df)
plt.plot(results2.fittedvalues, color='red')
plt.show()
```

```python
df['forecast']=results2.predict(start=len(df)-35, end=len(df)-10, dynamic=True)
df[['Prev Close', 'forecast']].plot(figsize=(20,15))
plt.show()
```

```python
from pandas.tseries.offsets import DateOffset
future_dates=[df.index[-1]+ DateOffset(weeks=x)for x in range(0, 52)]
```

```python
future_datest_df=pd.DataFrame(index=future_dates[1:], columns=df.columns)
future_datest_df.tail()
```

```python
future_df=pd.concat([df,future_datest_df])
future_df['forecast']=results2.predict(start=len(df), end=len(df)+52, dynamic=True)  
future_df[['Prev Close', 'forecast']].plot(figsize=(20, 15))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('Previous Prices', fontsize=28)
plt.title('Previous Price Forcasting upto September, 2022', fontsize=28)
plt.legend(loc='best', prop={'size': 18})
plt.show()
```
